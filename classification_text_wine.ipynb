```
from sklearn.datasets import load_wine

wine = load_wine()
wine
```

```
import pandas as pd

df = pd.DataFrame(wine["data"],columns=wine["feature_names"])
df["target"] = wine["target"]
df
```

```
df.corr()
```

```
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# 散點圖 ,參數hue 表示色調,換句話說就是你要呈現的資料標籤會自動標色。
# sns.pairplot(df,hue = "target")
plt.figure(figsize=(10,8))
sns.heatmap(df.astype("float").corr(),cmap="hot" ,annot=True)
```

```
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df.drop(["target"],axis=1),df["target"],test_size=0.1)
```

```
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(max_depth=3)
clf.fit(x_train,y_train)
```

```
from sklearn.tree import export_graphviz
import graphviz

g = export_graphviz(clf,out_file=None, special_characters=True, filled=True, feature_names=wine["feature_names"], class_names = wine["target_names"])
graph = graphviz.Source(g)
graph
```

```
test_result = clf.predict(x_test)

print("預測結果:",list(test_result))
print("實際結果:",list(y_test))
```

```
from sklearn.metrics import accuracy_score

pr = accuracy_score(y_test, test_result)
pr
```

```
from sklearn.metrics import confusion_matrix

M = confusion_matrix(y_test, test_result)
M
```
