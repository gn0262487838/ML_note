```

from sklearn.datasets import load_boston
boston = load_boston()
boston
```

```

import pandas as pd

df = pd.DataFrame(boston["data"],columns=boston["feature_names"])
df["target"] = boston["target"]
df
```

```

# pandas: .corr()
df.corr()
```

```

import matplotlib.pyplot as plt

# matplotlib
import seaborn as sns

# show,%matplotlib inline .show()
%matplotlib inline

# float,!!!
# pandasastype()

#  
plt.figure(figsize=(10,8))

# : cmap matplotlib colormap
# : annot 
sns.heatmap(df.astype("float").corr(),cmap="GnBu", annot=True)
```

```

from sklearn.model_selection import train_test_split
# goals
# train _test_split >>> (feature 90%, feature 10%, target 90%, target 10%)

# axis = 1 ,target
x_train, x_test, y_train, y_test = train_test_split(df.drop(["target"],axis=1),df["target"],test_size=0.1)
```

```
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor(max_depth=5)
# fit()
reg.fit(x_train,y_train)
```

```
from sklearn.tree import export_graphviz
import graphviz
#  filled graphviz
g = export_graphviz(reg,out_file = None,
                feature_names = boston["feature_names"], filled=True, special_characters=True)

# 
graph = graphviz.Source(g)
graph
```

```
test_result = reg.predict(x_test)
print(":",list(test_result))
print(":",list(y_test))

```

```
#  r2_score 
from sklearn.metrics import r2_score

print(":",r2_score(y_test,test_result))

```
